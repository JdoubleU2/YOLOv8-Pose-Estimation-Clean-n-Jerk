```python
from ultralytics import YOLO
```

```python
# Load and train YOLOv8 pose pre-trained model
model = YOLO('yolov8n-pose.pt')
model.train(data='/Users/jabinwade/Coding/ComputerVision/ComputerVision-FinalProject/data/Powerlifting-Clean and Jerk-.v2i.yolov8/data.yaml', epochs=100, imgsz=640)
model.save('powerlifting-pose.pt')
```

```python
# Evaluate on test set after training
metrics = model.val(data='Users/jabinwade/Coding/ComputerVision/ComputerVision-FinalProject/data/Powerlifting-Clean and Jerk-.v2i.yolov8/data.yaml', split='test')
print("Test set evaluation:", metrics)
```
