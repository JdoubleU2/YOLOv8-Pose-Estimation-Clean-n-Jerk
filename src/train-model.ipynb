```python
# Load and train YOLOv8 pose pre-trained model
model = YOLO('/content/powerlifting-pose.pt')
model.train(data='/content/drive/MyDrive/ComputerVision-FinalProject/data/Powerlifting-Clean and Jerk-.v2i.yolov8/data.yaml', epochs=150, imgsz=640)
model.save('powerlifting-pose.pt')
```

```python
# Evaluate on test set after training
metrics = model.val(data='/content/drive/MyDrive/ComputerVision-FinalProject/data/Powerlifting-Clean and Jerk-.v2i.yolov8/data.yaml', split='test')
print("Test set evaluation:", metrics)
```
